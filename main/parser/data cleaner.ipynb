-------------

Lets clean column by column

**Issues**:
- subject 1 has string values in the date column
- subject 1 misses temperature columns

```python
def clean(row):
    if isinstance(row, str):
        if '_' in row:
            day, month = row.split('_')
            if month == 'Dec':
                return pd.to_datetime('2021-12-{}'.format(day))
    return row
```

```python
for i, df in enumerate(dfs):
    dfs[i]['date'] = df['date'].apply(clean)
    dfs[i] = df.astype({'date': 'datetime64[ns]'})
    dfs[i] = df.set_index('date')
```

```python
dfs[0]['temperature'] = dfs[1]['temperature']
```

---------------
Select the valid dates


**Notes**:
- November, 26th, was the last day a surgical mask was worn. As we only measured the day partly, we can remove the values.
- December, 17th, was the last day of the experiment. As we only measured the day partly, we can remove the values.

```python
from datetime import datetime
```

```python
start_date = datetime(2021, 11, 18, 0, 0)
end_date = datetime(2021, 12, 17, 0, 0)
nov_26 = datetime(2021, 11, 26, 0, 0)
dec_17 = datetime(2021, 12, 17, 0, 0)
```

```python
for i, df in enumerate(dfs):
    temp = df.loc[start_date:end_date]
    dfs[i] = temp.drop(index=[nov_26, dec_17])
```

The only missing data seems to be the temperature. Only the room temperature is important. Luckily another subject recorded the room temperature. Thus, we can use the temperature recorded by the other subject.

Some columns have string values when it should be an integer.

```python
import math
def clean_acne(row):
    if isinstance(row, str):
        return 0
        
    return row
```

```python
spo2_cols = ['spo2_m1_r','spo2_m1_l','spo2_m2_r','spo2_m2_l','spo2_m3_r','spo2_m3_l']

for i, df in enumerate(dfs):
    # Most of the times room temperature was 20 degrees according to subject 4, thus we fill the rest with 20
    dfs[i]['temperature'] = 20
    dfs[i]['sleep'] = df['sleep'].astype('int')
    dfs[i]['acne'] = df['acne'].apply(clean_acne)
    dfs[i][spo2_cols] = df[spo2_cols].applymap(lambda x: x / 100 if x > 1 else x)
```

Some specifications of the column types:
        
        date           datetime64[ns]
        spo2_...       float, where0 < x < 1 and is a percentage
        temperature    int, degrees celsius
        masktype       string
        acne           int
        stress         int, level of stress between 1 and 10
        sleep          int, amount of sleep in hours

```python
types = {
    'masktype': 'object',
    'acne': 'int',
    'spo2_m1_r': 'float',
    'spo2_m1_l': 'float',
    'spo2_m2_r': 'float',
    'spo2_m2_l': 'float',
    'spo2_m3_r': 'float',
    'spo2_m3_l': 'float',
    'stress': 'int',
    'sleep': 'int',
    'temperature': 'int'
}
```

```python
for i, df in enumerate(dfs):
    dfs[i] = df.astype(types)
```
